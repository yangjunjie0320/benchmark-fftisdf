```python
import numpy, scipy

data = numpy.loadtxt("save-vjk-nz.out", delimiter=",")
ene_ref = -43.191002

dd = []
for i in range(data.shape[0]):
    ene_err = abs(data[i, 5] - ene_ref) / 8
    dd.append([data[i, 0], data[i, 1], data[i, 2], data[i, 3], data[i, 4], ene_err])

# sort dd by ene_err
dd.sort(key=lambda x: x[5])

import sys
numpy.savetxt(sys.stdout, dd, fmt="% 6.2e", delimiter=",")
```
